```python
NoFilter = pd.read_csv('lasernf.csv',delimiter=",",names = ['wave','sig'],skiprows=33,skipfooter=1,engine='python')
Filters = pd.read_csv('laserLLF-SPF.csv',delimiter=",",names = ['wave','sig'],skiprows=33,skipfooter=1,engine='python')
```

```python
NF =savgol_filter(NoFilter['sig'], 5, 3)
FF =savgol_filter(Filters['sig'], 5, 3)
```

```python
plt.figure(figsize=(10,10))
plt.plot(NoFilter['wave'],NF,label='Laser - unfiltered',color='dimgray',linewidth=2)
plt.plot(Filters['wave'],FF,label='Laser - filtered',color='b',linewidth=2)
plt.semilogy()

plt.legend( loc='upper left',fontsize=26,shadow=True, fancybox=True)
plt.xlabel(r'Wavelength / nm', fontsize=28)
plt.ylabel('Intensity (arb. units)', fontsize=28)

plt.xticks(fontsize = 26)
plt.yticks(fontsize = 26)
plt.xticks([450,460,470,480,490,500],ha='left')
plt.tick_params(top="off",right='off')
plt.tick_params(axis='x', direction='out',width=2,length=5)
plt.tick_params(axis='y',width=2,length=5)


#plt.title('Argon Ion Spectrum',fontsize=32)
plt.xlim(450,510)
#plt.ylim(0,.99)
plt.tight_layout() 
plt.ylim(0.8e-3,1.1)

#plt.savefig('/Users/austinmcdonald/Desktop/Plots/paper8-16/Arionspeclog.pdf', format='pdf')

plt.show()
```
